```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Simple Low-Pass Filter
def low_pass_filter(data, alpha=0.0):
    filtered_data = np.copy(data)
    for i in range(1, len(filtered_data)):
        filtered_data[i] = alpha * filtered_data[i] + (1 - alpha) * filtered_data[i - 1]
    return filtered_data

# Simple High-Pass Filter
def high_pass_filter(data, alpha=0.1):
    filtered_data = np.copy(data)
    for i in range(1, len(filtered_data)):
        filtered_data[i] = alpha * (filtered_data[i] - filtered_data[i - 1]) + filtered_data[i - 1]
    return filtered_data

# Load the audio file
y, sr = librosa.load(r"C:\Users\kalin\Desktop\IIR_Filter_Basics\1.mp3", sr=None) # Replace with your file path

# Apply Filters
lpf_samples = low_pass_filter(y)
hpf_samples = high_pass_filter(y)

# Plotting
plt.figure(figsize=(15, 6))

plt.subplot(3, 1, 1)
librosa.display.waveshow(y, sr=sr, alpha=0.5, label='Original', color='blue')
plt.title('Original Audio Signal')
plt.legend()

plt.subplot(3, 1, 2)
librosa.display.waveshow(lpf_samples, sr=sr, alpha=0.5, label='Low-Pass Filtered', color='green')
plt.title('Low-Pass Filtered Signal')
plt.legend()

plt.subplot(3, 1, 3)
librosa.display.waveshow(hpf_samples, sr=sr, alpha=0.5, label='High-Pass Filtered', color='red')
plt.title('High-Pass Filtered Signal')
plt.legend()

plt.tight_layout()
plt.show()
```
