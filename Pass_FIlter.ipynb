```python
import numpy as np
import matplotlib.pyplot as plt

class Filter:
    def __init__(self, filter_value=0.9):
        self.filter_value = filter_value
        self.filtered_signal_low = 0
        self.filtered_signal_high = 0
        self.filter_helper = 0
    
    def low_pass(self, input_signal):
        self.filtered_signal_low = input_signal * (1- self.filter_value) + self.filtered_signal_low * self.filter_value
        return self.filtered_signal_low
    
    def high_pass(self, input_signal):
        self.filter_helper = input_signal * self.filter_value + self.filter_helper * (1- self.filter_value)
        self.filtered_signal_high = input_signal - self.filter_helper
        return self.filtered_signal_high

def generate_signal(time):
    # Erzeugt ein Beispiel-Signal: eine Sinuswelle mit Rauschen
    base_signal = np.sin(time*100) + np.sin(time) + np.sin(time*23)
    noise = np.random.normal(0, 0.5, len(time))
    noise= 0
    return base_signal + noise

# Zeitachse erstellen
time = np.linspace(0, 10, 1000)
# Beispiel-Signal erzeugen
input_signal = generate_signal(time)

# Filter-Instanz erstellen
my_filter = Filter(filter_value=0.9)

# Filtern des Signals
low_passed = [my_filter.low_pass(val) for val in input_signal]
high_passed = [my_filter.high_pass(val) for val in input_signal]

# Plotten der Ergebnisse
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(time, input_signal, label='Original Signal')
plt.title('Original Signal')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(time, low_passed, label='Low-pass Filtered', color='orange')
plt.title('Low-pass Filtered')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(time, high_passed, label='High-pass Filtered', color='green')
plt.title('High-pass Filtered')
plt.legend()

plt.tight_layout()
plt.show()
```
